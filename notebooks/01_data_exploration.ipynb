# CardioForecaster — Data Exploration

Inspect the structure of synthetic ICU episodes and the MIMIC-III waveform
channels we forecast. The synthetic generator is schema-compatible with the
MIMIC loader so the rest of the pipeline does not care which source produced
the data.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path('..').resolve() / 'backend'))

import matplotlib.pyplot as plt
from app.data.synthetic import generate_episode, CHANNELS
```

```python
signals, labels, patient = generate_episode(n_minutes=480, deterioration=True, seed=7)
print('signal shape', signals.shape)
print('deterioration window:', labels.sum(), 'samples')
print('patient phenotype:', patient)
```

```python
fig, axes = plt.subplots(len(CHANNELS), 1, figsize=(12, 9), sharex=True)
for ax, name, idx in zip(axes, CHANNELS, range(len(CHANNELS))):
    ax.plot(signals[:, idx], lw=0.6)
    ax.set_ylabel(name)
    ax.fill_between(range(len(labels)), signals[:, idx].min(), signals[:, idx].max(),
                    where=labels.astype(bool), alpha=0.15, color='red')
axes[-1].set_xlabel('seconds')
fig.suptitle('Synthetic ICU episode with deterioration overlay')
fig.tight_layout()
```
